Final Projesi: Dijitalleşmenin Toplumsal Etkileri

Hazırlayanlar: Ronya Hsso, Merve Şen, Nisanur Ay, Pelin Ay

MP1: Problem ve Hipotez
Dijitalleşme, teknolojik gelişmelerin işletme ve toplum hayatına hızla entegre edilmesi sürecidir. Bu konuyu, güncel olması ve bilinçli gençler olma amacımız doğrultusunda seçtik.






Hipotez: Yapay zeka ve otomasyonun iş süreçlerine entegrasyonu, düşük beceri gerektiren mesleklerde istihdamı azaltırken, yüksek beceri gerektiren yeni roller yaratarak iş gücü piyasasında yapısal bir dönüşüme neden olmaktadır.


Bağımsız Değişken: Dijital teknolojilerin kullanım yoğunluğu.


Bağımlı Değişken: Toplam istihdam düzeyi ve yeni mesleklerin oranı.

MP2: Literatür Taraması
Literatür çalışmalarımızda dijitalleşmenin hem bireysel hem de kurumsal etkilerini inceledik:


Dijital Hegemonya: Dijital araçların bireylerin davranışlarını şekillendiren görünmez bir gözetim ve denetim sistemi kurduğu görülmektedir.


Dijital Talep: Sağlık gibi sektörlerde fiziksel dilek kutularının yerini hız, kolaylık ve erişilebilirlik sağlayan dijital platformlar almıştır.


İstihdam Etkisi: Otomasyon verimliliği artırıp yeni meslekler (veri bilimci, siber güvenlik uzmanı vb.) yaratsa da, orta beceri düzeyindeki işlerde kayıplara ve gelir eşitsizliğine yol açabilmektedir.

MP3: Veri Toplama
Bu projede veriler TÜİK (2024) Çocuklarda Bilişim Teknolojileri Kullanım Araştırması sonuçlarından elde edilmiştir.




Yöntem: Açık veri setleri ve Web Scraping yöntemleri kullanılmıştır.



Bulgular: Çocukların internet kullanım oranı 2021'de %82,7 iken 2024'te %91,3'e yükselmiştir.

MP4: Veri Analizi ve Görselleştirme
1. İnternet Kullanım Amaçları (Bar Chart / Sütun Grafiği)
Bu grafik, çocukların interneti hangi faaliyetler için kullandığını göstermektedir.

Bulgu: En yüksek kullanım %83,9 ile video izleme, en düşük ise %13,2 ile e-posta iletişimidir.

2. Yaş Grupları Arasındaki İlişki (Scatter Plot / Serpilme Grafiği)
Bu grafik, 6-10 yaş ve 11-15 yaş gruplarının kullanım yüzdelerini karşılaştırır.

Analiz: Noktaların doğrusal bir hat üzerinde dizilmesi, iki grup arasında mükemmel bir pozitif korelasyon olduğunu kanıtlar.

Sonuç: Bir faaliyet bir grupta popülerse diğerinde de popülerdir; ancak 11-15 yaş grubu her zaman daha yüksek kullanım oranına sahiptir.


Kullanım Dağılımı (Histogram)
İnternet faaliyetlerinin yoğunluğunu gösterir.

Bulgu: Kullanım motivasyonları "yüksek katılım" (eğlence) ve "düşük katılım" (resmi işler) olarak iki uçta toplanmıştır.

Yaş Grupları Korelasyonu (Heatmap)
Farklı yaş grupları (6-10 ve 11-15) arasında kullanım amaçları açısından neredeyse mükemmel bir uyum gözlemlenmiştir. Eğer bir faaliyet küçük grupta popülerse, büyük grupta da popülerdir.

MP5: Raporlama ve Öneri
Sonuç
Yaş ilerledikçe dijital faaliyetlere katılım frekansı doğrusal bir artış göstermektedir ancak faaliyetlerin türü yaş grupları arasında homojen bir dağılım sergilemektedir.


Öneriler

Eğitim: Okul öncesinden itibaren dijital okuryazarlık ve siber güvenlik eğitimleri müfredata entegre edilmelidir.


Beceri Gelişimi: Otomasyonun taklit edemeyeceği yaratıcılık, eleştirel düşünme ve duygusal zeka gibi "insan odaklı" beceriler desteklenmelidir.


Strateji: Yaşa göre farklı içerik kategorileri yerine, mevcut yüksek etkileşimli alanların yaşa göre ölçeklendirilmiş versiyonlarına yatırım yapılmalıdır.
